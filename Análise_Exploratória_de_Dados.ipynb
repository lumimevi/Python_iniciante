*(b) Construa a distribuição de frequência e o histograma. Use 5 intervalos de classe*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def gerar_histograma(dados, n_classes):
    freq, bins = np.histogram(dados, bins=n_classes)
    freq_relativa = freq / len(dados)
    classe_labels = [f"{round(bins[i], 1)} - {round(bins[i+1], 1)}" for i in range(len(bins)-1)]

    tabela_freq = pd.DataFrame({
        "Intervalo de Classe": classe_labels,
        "Frequência": freq,
        "Frequência Relativa": freq_relativa
    })

    plt.figure(figsize=(8, 5))
    plt.hist(dados, bins=bins, edgecolor='black')
    plt.title(f"Histograma com {n_classes} Intervalos de Classe")
    plt.xlabel("Valor")
    plt.ylabel("Frequência")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()

    return tabela_freq

#  5 intervalos
tabela_5_classes = gerar_histograma(dados, 5)
print("\nDistribuição de Frequência (5 intervalos):")
print(tabela_5_classes)
```

*(c) Construa a distribuição de frequência e o histograma, agora com 8 intervalos de classe.*

```python
#  8 intervalos
tabela_8_classes = gerar_histograma(dados, 8)
print("\nDistribuição de Frequência (8 intervalos):")
print(tabela_8_classes)
```

```python
import numpy as np

# Dados
sambaqui = [30.61, 28.89, 32.21, 24.25, 25.63,
    42.88, 36.22, 28.86, 22.56, 22.92,
    27.94, 41.45, 42.59, 15.25, 33.29]

manguezal = [25.34, 25.67, 17.64, 33.97, 11.13,
    9.49, 16.92, 12.91, 14.05, 14.88,
    19.17, 21.60, 20.01, 19.81, 16.22]
```

*a) calcule a média e a mediana para cada um dos locais. Onde houve maior crescimento?*

*b)  Calcule o  Q1 e  o Q3 para  cada  um  dos  locais. Explique o signiﬁcado  destes  números.*

*c)  Calcule a variância para cada um dos locais.*


```python
# Função para calcular estatísticas
def estatisticas(nome, dados):
    media = np.mean(dados)
    mediana = np.median(dados)
    q1 = np.percentile(dados, 25)
    q3 = np.percentile(dados, 75)
    variancia = np.var(dados, ddof=1)

    print(f"\n📍 Estatísticas para {nome}:")
    print(f"Média: {media:.2f} g")
    print(f"Mediana: {mediana:.2f} g")
    print(f"1º Quartil (Q1): {q1:.2f} g")
    print(f"3º Quartil (Q3): {q3:.2f} g")
    print(f"Variância: {variancia:.2f} g²")

# Resultados para cada local
estatisticas("Sambaqui", sambaqui)
estatisticas("Manguezal", manguezal)
# Comparação de crescimento
media_sambaqui = np.mean(sambaqui)
media_manguezal = np.mean(manguezal)

if media_sambaqui > media_manguezal:
    print("\n✅ Houve maior crescimento em *Sambaqui* (maior peso médio da carne).")
else:
    print("\n✅ Houve maior crescimento em *Manguezal* (maior peso médio da carne).")
```


**(a) Média e Mediana:**

-Média indica o valor médio de peso da carne.

-Mediana mostra o valor central da amostra ordenada.

Comparar as duas permite verificar crescimento médio e detectar assimetrias.

**(b) Q1 e Q3:**

Q1 (25%): 25% dos mexilhões pesam menos que este valor.

Q3 (75%): 75% dos mexilhões pesam menos que este valor.

Q1 e o Q3, ajudam a entender a dispersão central e identificar a variabilidade entre os pesos.

**(c) Variância:**

Mede o grau de dispersão dos pesos em relação à média.

Uma variância maior significa que os pesos estão mais espalhados.

#Questão 5#

*Temos abaixo a distribuição de frequências dos pesos de uma amostra de 50 alunos.*



image.png

*a)	determinar a média*

```python
import numpy as np

# Frequências (f) e Pontos médios (xi)
frequencias = np.array([4, 16, 13, 9, 5, 3])
pontos_medios = np.array([42.5, 47.5, 52.5, 57.5, 62.5, 67.5])
# Total de alunos
n = np.sum(frequencias)

# (a) Média ponderada
media = np.sum(frequencias * pontos_medios) / n
print(f"(a) Média: {media:.2f} kg")
```

*b)	determinar a variância*

```python
# (b) Variância ponderada
variancia = np.sum(frequencias * (pontos_medios - media) ** 2) / n
print(f"(b) Variância: {variancia:.2f} kg²")
```

Indica o grau de dispersão dos dados em torno da média.

*c)	determinar o coeficiente de variação*

```python
# (c) Coeficiente de variação
desvio_padrao = np.sqrt(variancia)
coef_var = (desvio_padrao / media) * 100
print(f"(c) Coeficiente de Variação: {coef_var:.2f}%")
```

Coeficiente de Variação, mede a dispersão em relação à média, útil para comparar variabilidade entre diferentes conjuntos de dados

*d)	a distribuição é simétrica?*

```python
#Simetria (baseada na forma da frequência)
# A frequência cresce até 50kg (pico) e decresce depois, indicando simetria aproximada
print(f" A distribuição é aproximadamente simétrica? {'Sim' if abs(media - 52.5) < 1 else 'Não'}")
```

*e)	a distribuição é mesocúrtica?*



```python
# Distribuição com formato de sino e sem caudas longas geralmente é mesocúrtica
from scipy.stats import kurtosis
# Aproxima os dados reais com base nas frequências
dados_aproximados = np.repeat(pontos_medios, frequencias)
curtose = kurtosis(dados_aproximados, fisher=True)  # Curtose de Fisher: 0 = mesocúrtica

print(f"Curtose de Fisher: {curtose:.2f} → {'Mesocúrtica' if abs(curtose) < 1 else 'Não mesocúrtica'}")
```

Se a curtose de Fisher estiver próxima de 0, a distribuição é mesocúrtica (sem caudas pesadas ou picos acentuados)

Como as calorias queimadas após o exercício contribuem para a perda de peso e têm outras consequências, é importante entender esse processo. O estudo “Effect of Weight Training Exercise and Treadmill Exercise on Post-Exercise Oxygen Consumption” (Medicine and Science in Sports and Exercise, 1998, p. 518-522) relatou os dados a seguir de um estudo em que o consumo de oxigênio (litros) foi medido continuamente por 30 minutos para cada um dos 15 indivíduos após um exercício de levantamento de peso e após um exercício em esteira rolante.*

*a.Construa um boxplot comparativo das observações de levantamento de peso e de esteira rolante e comente as características encontradas*

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Dados
levantamento = [14.6, 14.4, 19.5, 24.3, 16.3, 22.1, 23.0, 23.2, 18.7, 18.5, 19.0, 15.9, 17.0, 19.1, 19.6]
esteira =      [11.3, 20.8, 23.6, 5.3, 10.3, 12.6, 9.1, 10.3, 4.4, 15.2, 10.1, 19.6, 2.6, 16.6, 22.4]

# Diferenças
diferencas = [x - y for x, y in zip(levantamento, esteira)]

#Boxplot comparativo
plt.figure(figsize=(8, 5))
sns.boxplot(data=[levantamento, esteira], palette="Set2")
plt.xticks([0, 1], ['Levantamento de peso', 'Esteira rolante'])
plt.ylabel("Consumo de Oxigênio (litros)")
plt.title("Boxplot Comparativo: Levantamento vs Esteira")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
